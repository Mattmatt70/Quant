```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, LSTM
import yfinance as yf

# Liste des tickers
tickers = ['AAPL', 'MSFT', 'MC.PA', 'RMS.PA']

for ticker in tickers:
    print("Processing", ticker)
    # Télécharger les données boursières du ticker actuel
    data = yf.download(ticker, start='2019-01-01', interval='1d')

    # Afficher les premières lignes des données
    print(data.head())

    # Prétraitement des données
    data["Close"] = pd.to_numeric(data["Close"], errors='coerce').fillna(method="ffill")
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(data[['Close']])

    feature_length = 100

    def Create_Features_and_Targets(data, feature_length):
        X = list()
        Y = list()
        for i in range(len(data) - feature_length - 1):
            X.append(data[i:(i + feature_length), 0])
            Y.append(data[i + feature_length, 0])
        X = np.array(X)
        Y = np.array(Y)
        return X, Y

    X_train, y_train = Create_Features_and_Targets(scaled_data, feature_length)

    X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
    X_train.shape, y_train.shape

    testData = data.iloc[:, 3:4]
    y_real = testData.iloc[feature_length + 1:, 0:].values
    x_test = testData.iloc[:, 0:].values

    x_test = scaler.transform(x_test)
    x_test, y_test = Create_Features_and_Targets(x_test, feature_length)

    x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], 1))

    # Création du modèle LSTM
    model = Sequential()
    model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], 1)))
    model.add(Dropout(0.2))
    model.add(LSTM(units=50, return_sequences=False))
    model.add(Dropout(0.2))
    model.add(Dense(units=1))

    # Compilation du modèle
    model.compile(optimizer='adam', loss='mean_squared_error')

    # Entraînement du modèle
    model.fit(X_train, y_train, epochs=100, batch_size=12)

    # Prédire les prix des actions
    y_pred = model.predict(x_test)
    predicted_price = scaler.inverse_transform(y_pred)

    # Afficher les prédictions
    plt.plot(y_real, color='red', label='Actual')
    plt.plot(predicted_price, color='green', label='Predicted')
    plt.xlabel('Time')
    plt.ylabel('Stock')
    plt.legend()
    plt.title(f'Stock Price Prediction for {ticker}')

    # Prédire la clôture pour demain
    last_data = np.expand_dims(x_test[-1], axis=0)  # Utiliser les données du dernier jour
    tomorrow_price = model.predict(last_data)
    tomorrow_price = scaler.inverse_transform(tomorrow_price)
    plt.text(1.01, 0.5, f'Predicted closing price for tomorrow: {tomorrow_price[0][0]}', fontsize=10, ha='left', va='center', transform=plt.gca().transAxes)

    plt.show()
```
