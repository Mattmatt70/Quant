```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Générer des données factices pour illustrer le concept
strikes = np.linspace(3000, 3500, num=10)
maturities = np.linspace(0.01, 1, num=10)
strikes, maturities = np.meshgrid(strikes, maturities)
implied_volatilities = 0.2 + 0.1 * np.random.randn(*strikes.shape)

# Création de la surface de volatilité implicite en 3D
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(strikes, maturities, implied_volatilities, cmap='viridis')
ax.set_xlabel('Strike')
ax.set_ylabel('Maturity')
ax.set_zlabel('Implied Volatility')
ax.set_title('Implied Volatility Surface for S&P 500 Options (Dummy Data)')
plt.show()
```
