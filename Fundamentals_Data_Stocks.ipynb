```python
#!pip install urllib3
#!pip install bs4

# For data manipulation
import pandas as pd
from urllib.request import urlopen, Request

# To extract fundamental data
from bs4 import BeautifulSoup

def fundamental_metric(soup, metric):
    return soup.find(text = metric).find_next(class_='snapshot-td2').text

def get_fundamental_data(df):
    for symbol in df.index:
        try:
            url = ("http://finviz.com/quote.ashx?t=" + symbol.lower())
            req = Request(url=url,headers={'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64; rv:20.0) Gecko/20100101 Firefox/20.0'})
            response = urlopen(req)
            soup = BeautifulSoup(response)
            for m in df.columns:
                df.loc[symbol,m] = fundamental_metric(soup,m)
        except Exception as e:
            print (symbol, 'not found')
    return df

stock_list = ['AMZN','GOOG','PG','KO','IBM','DG','XOM','KO','PEP','MT','NL','LPL']

metric = ['Price',
'Change',
'52W High',
'52W Low',
'Perf Week',
'Perf Month',
'Perf Quarter',
'Perf Half Y',
'Perf Year',
'Perf YTD',
'P/B',
'P/E',
'Forward P/E',
'PEG',
'P/S',
'P/C',
'P/FCF',
'Debt/Eq',
'EPS (ttm)',
'Debt/Eq',
'LT Debt/Eq',
'Market Cap',
'RSI (14)',
'Target Price',
'ATR (14)',
'Volume',
'Beta',
'Recom',
'ROA',
'ROE',
'ROI',
'Gross Margin',
'Oper. Margin',
'Profit Margin',
'Payout'
]

df = pd.DataFrame(index=stock_list,columns=metric)
df = get_fundamental_data(df)
df
```

```python
def convert_market_cap(market_cap):
    if 'B' in market_cap:
        return float(market_cap.replace('B', '')) * 1e9
    elif 'M' in market_cap:
        return float(market_cap.replace('M', '')) * 1e6
    else:
        return float(market_cap)

# Appliquer la fonction à la colonne 'Market Cap'
df['Market Cap'] = df['Market Cap'].apply(convert_market_cap)
```

```python
df['Change'] = df['Change'].str.replace('%', '')
df['Volume'] = df['Volume'].str.replace(',', '.')
df['52W High'] = df['52W High'].str.replace('%', '')
df['52W Low'] = df['52W Low'].str.replace('%', '')
df['Perf Week'] = df['Perf Week'].str.replace('%', '')
df['Perf Month'] = df['Perf Month'].str.replace('%', '')
df['Perf Quarter'] = df['Perf Quarter'].str.replace('%', '')
df['Perf Half Y'] = df['Perf Half Y'].str.replace('%', '')
df['Perf Year'] = df['Perf Year'].str.replace('%', '')
df['Perf YTD'] = df['Perf YTD'].str.replace('%', '')
df['ROA'] = df['ROA'].str.replace('%', '')
df['ROE'] = df['ROE'].str.replace('%', '')
df['ROI'] = df['ROI'].str.replace('%', '')
df['Gross Margin'] = df['Gross Margin'].str.replace('%', '')
df['Oper. Margin'] = df['Oper. Margin'].str.replace('%', '')
df['Profit Margin'] = df['Profit Margin'].str.replace('%', '')
df['Payout'] = df['Payout'].str.replace('%', '')
df = df.apply(pd.to_numeric, errors='coerce')
df
```
