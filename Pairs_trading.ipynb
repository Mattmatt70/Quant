```python
import yfinance as yf
import datetime
import statsmodels.api as sm
import matplotlib.pyplot as plt
import numpy as np

# Symboles des paires de devises
symbols = ['EURUSD=X', 'GBPUSD=X']

# Plage de dates souhaitée
start_date = '2022-09-19'  # Date de début il y a 5 ans
end_date = '2023-09-19'    # Date de fin actuelle

# Récupérer les données depuis Yahoo Finance
data = yf.download(symbols, start=start_date, end=end_date)

# Afficher les données de l'EUR/USD et du GBP/USD sur le même graphique
plt.figure(figsize=(15, 7))
plt.plot(data.index, data[('Adj Close', 'EURUSD=X')], label='EUR/USD', color='blue')
plt.plot(data.index, data[('Adj Close', 'GBPUSD=X')], label='GBP/USD', color='green')
plt.legend(['EUR/USD', 'GBP/USD'])
plt.title('Cours de l\'EUR/USD et du GBP/USD')
plt.xlabel('Date')
plt.ylabel('Cours')
plt.show()

# Calculer la corrélation entre l'EUR/USD et le GBP/USD
correlation = data[('Adj Close', 'EURUSD=X')].corr(data[('Adj Close', 'GBPUSD=X')])
print("Corrélation entre EUR/USD et GBP/USD sur la période spécifiée :", correlation)

# Extraire les séries temporelles de clôture (Adj Close) pour l'EUR/USD et le GBP/USD
S1 = data[('Adj Close', 'EURUSD=X')]
S2 = data[('Adj Close', 'GBPUSD=X')]

# Calculer la cointégration
coint_result = sm.tsa.coint(S1, S2)

# Récupérer le score de cointégration, la p-value et les vecteurs de cointégration
score = coint_result[0]
pvalue = coint_result[1]
coint_vectors = coint_result[2]

# Calculer le ratio
ratios = S1 / S2

# Créer un graphique du ratio
plt.figure(figsize=(15, 7))
plt.plot(ratios.index, ratios, color='blue')
plt.axhline(ratios.mean(), color='red', linestyle='--', label='Moyenne')
plt.legend(['Ratio', 'Moyenne'])
plt.title('Ratio entre EUR/USD et GBP/USD')
plt.xlabel('Date')
plt.ylabel('Ratio')
plt.show()

print('Score de cointégration : {:.4f}'.format(score))
print('P-value : {:.4f}'.format(pvalue))
print('Mean ratio : {:.4f}'.format(ratios.mean()))

# Normaliser le ratio (calcul du Z-score)
def zscore(series):
    return (series - series.mean()) / np.std(series)

# Créer un graphique du Z-score du ratio avec des lignes pour les seuils de -3, -2, -1, +1, +2 et +3 en solide
plt.figure(figsize=(15, 7))
plt.plot(zscore(ratios).index, zscore(ratios), color='blue')
plt.axhline(zscore(ratios).mean(), color='black', linestyle='--', label='Moyenne Z-score')
plt.axhline(3.0, color='red', linestyle='-', label='Z-score 3.0', linewidth=2)
plt.axhline(-3.0, color='green', linestyle='-', label='Z-score -3.0', linewidth=2)
plt.axhline(2.0, color='red', linestyle='-', label='Z-score 2.0')
plt.axhline(-2.0, color='green', linestyle='-', label='Z-score -2.0')
plt.axhline(1.0, color='red', linestyle='--', label='Z-score 1.0')
plt.axhline(-1.0, color='green', linestyle='--', label='Z-score -1.0')
plt.legend(['Z-score', 'Moyenne Z-score', 'Z-score 3.0', 'Z-score -3.0', 'Z-score 2.0', 'Z-score -2.0', 'Z-score 1.0', 'Z-score -1.0'])
plt.title('Z-score du Ratio entre EUR/USD et GBP/USD')
plt.xlabel('Date')
plt.ylabel('Z-score')
plt.show()
```
